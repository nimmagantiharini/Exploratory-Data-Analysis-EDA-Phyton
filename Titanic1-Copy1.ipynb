```python
import pandas as pd  
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
gender_df = pd.read_csv("gender file titanic 1.csv")
```

```python
test_df   = pd.read_csv("Test file titanic 2.csv")
```

```python
train_df  = pd.read_csv("Train file titanic 3.csv")
```

```python
sns.countplot(x='Survived', data=train_df)
plt.title("Survival Distribution")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=train_df, palette=['black', 'red'])
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.histplot(train_df['Age'].dropna(), kde=True, bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
numeric_cols = ['Survived', 'Age', 'Fare', 'Pclass', 'SibSp', 'Parch']
```

```python
sns.pairplot(train_df[numeric_cols], hue="Survived", diag_kind="kde", palette="coolwarm")
plt.suptitle("Pairplot of Titanic Dataset", y=1.02)
plt.show()
```

```python
numeric_df = train_df[['Survived','Pclass','Age','SibSp','Parch','Fare']]
```

```python
corr = numeric_df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Correlation Heatmap - Titanic Dataset", fontsize=14)
plt.show()
```

```python
sns.histplot(train_df['Fare'], kde=True, bins=40)
plt.title("Fare Distribution")
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=train_df)
plt.title("Fare by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train_df, palette="coolwarm", alpha=0.7)
plt.title("Age vs Fare by Survival")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.stripplot(x='Sex', y='Age', hue='Survived', data=train_df, 
              jitter=True, dodge=True, alpha=0.6, palette="coolwarm")
plt.title("Age distribution by Gender and Survival")
plt.show()
```
